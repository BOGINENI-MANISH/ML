```python
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv('tested.csv')
df.drop('Fare', axis=1, inplace=True)

null_counts = df.isnull().sum()


plt.figure(figsize=(10, 6))
null_counts.plot(kind='bar', color='red', alpha=0.7)
plt.title('Number of Null Values per Column in Titanic Dataset')
plt.xlabel('Columns')
plt.ylabel('Count of Null Values')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()


plt.show()


df['Age'].fillna(df['Age'].mean(), inplace=True)


df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)


print(df.isnull().sum())

```

```python

import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv('tested.csv')
df.drop('Fare', axis=1, inplace=True)

null_counts = df.isnull().sum()

plt.figure(figsize=(10, 6))
# Directly create scatter plot using null_counts index and values
plt.scatter(null_counts.index, null_counts.values, color='red', alpha=0.7, linestyle='--')
plt.title('Number of Null Values per Column in Titanic Dataset')
plt.xlabel('Columns')
plt.ylabel('Count of Null Values')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()


df['Age'].fillna(df['Age'].mean(), inplace=True)


df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)


print(df.isnull().sum())
```
