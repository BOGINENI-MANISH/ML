```python
import numpy as np

arr = np.array([[1,2,3],
               [4,5,6]])
print("type of array: ",type(arr))
print("dimentions of array: ",arr.ndim)
print("shape of the array: ",arr.shape)
print("size of the array: ",arr.size)
print("array stores type of elements: ",arr.dtype)
```

```python
import numpy as np
a= np.array([[1,2,3],[4,5,6]],dtype='float')
print(" listed using passed array: ",a)

b=np.array((1,2,3))
print("\n array created using passed tupel: \n",b)

c=np.zeros((3,4))
print("\n an array intialised iwth all zeros:\n ",c)

d=np.full((3,3),6,dtype='complex')
print("\n an array instialised with all 6s.array type is complex:\n",d)

e=np.random.random((2,2))
print("\n an random array:\n",e)

f=np.arange(0,30,5)
print("\n a seqential array which steps of 5:\n",f)

g=np.linspace(0,5,10)
print("\n a sequential array with 10 values between 0 and 5:\n",g)

arr=np.array([[1,2,3,4],
              [5,3,4,2],
              [1,2,0,1]])
newarr= arr.reshape(2,2,3)
print("\n original array: ",arr)
print("\n reshaped array: ",newarr)

arr = np.array([[1,2,3],[4,5,6]])
flarr=arr.flatten()

print("\n original array: ",arr)
print("\n flatten array: ",flarr)
```

```python
import tensorflow as tf

node1= tf.constant(3,dtype=tf.int32)
node2= tf.constant(5,dtype=tf.int32)
node3= tf.add(node1,node2)

sess=tf.compat.v1.Session()

print("sum of node1 and node 2 is:",sess.run(node3))
sess.close()

```
